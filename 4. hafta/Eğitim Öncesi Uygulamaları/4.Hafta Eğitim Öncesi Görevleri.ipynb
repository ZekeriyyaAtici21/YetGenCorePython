# Problem 3

Kullanıcıya hangi öğrencinin notlarını görmek istediğini sorun. Öğrencinin notu görüntülendiğinde program sonunda şöyle bir çıktı elde etmelisiniz:

```

                              2.Sınav Notu:66

````python
name = input("Öğrencinin ismini giriniz : ")
sinavList1 = list(("46" , "65" , "85"))
sinavList2 = list(("76" , "96" , "55"))
sinavList3 = list(("60" , "100", "88"))
ogrenciDict = {"Hasan" :("\n1.Sinav Notu : "+ sinavList1[0]+ "\n2.Sinav Notu : "+ sinavList2[0]+ "\n3.Sinav Notu :"+sinavList3[0]) , "Ahmet" : (" 1.Sinav Notu : "+ sinavList1[1]+ "\n 2.Sinav Notu  "+ sinavList2[1]+ "\n3.Sinav Notu "+sinavList3[1]), "Merve" :(" 1.Sinav Notu : "+ sinavList1[2]+ "\n 2.Sinav Notu  "+ sinavList2[2]+ "\n3.Sinav Notu "+sinavList3[2])}

print(f"{name} isimli öğrencinin " , ogrenciDict[name])

if name.upper() == "AHMET":
    Ort = (int(sinavList1[1]) + int(sinavList2[1]) + int(sinavList3[1])) /3 
    print("\n Not Ortalaması : " , round(Ort,2))
elif name.upper() == "MERVE":
    Ort = (int(sinavList1[2]) + int(sinavList2[2]) + int(sinavList3[2])) /3 
    print("\n Not Ortalaması : " , round(Ort,2))
elif name.upper() =="HASAN":
    Ort = (int(sinavList1[0]) + int(sinavList2[0]) + int(sinavList3[0])) /3 
    print("\n Not Ortalaması : " , round(Ort,2))
else:
    print("Bu öğrenci kayıtlı değildir.")
````
