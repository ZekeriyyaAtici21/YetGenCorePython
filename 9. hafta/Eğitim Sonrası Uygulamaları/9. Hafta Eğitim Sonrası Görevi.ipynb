# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import time
import random


class commander():
    def __init__(self, tvDurum = "Kapalı", ses = 0, kanallarListesi = ["FOX"], kanal = "FOX"):
        self.tvDurum = tvDurum
        self.ses = ses
        self.kanallarListesi = kanallarListesi
        self.kanal = kanal
    
    def tvAcmak (self):

        if (self.tvDurum =="Açık" ):
            print("Televizyon açık durumda.")
        else:
            print("Televizyon açılıyor.")
            self.tvDurum = "Açık"
    
    def tvKapat(self):

        if ( self.tvDurum == "Kapalı"):
            print("Televiyon kapalı durumda.")
        else:
            print("Televizyon Kapatılıyor.")
            self.tvDurum = "Kapalı"
    
    def sesAyar(self):

        while 1:

            cevap = input("Sesi azalt tusu : '<'\nSesi artır tusu: '>'\nÇıkıs butonu:'x'")

            if cevap == "<":
                if(self.ses != 0):
                    self.ses -= 1
                    print("Ses Seviyesi: ",self.ses)
            elif cevap == ">":
                if self.ses != 20:
                    self.ses += 1
                    print("Ses level:",self.ses)
            else:
                print("Ses değiştirildi.")
    
    def ekle(self,kanalİsmi):

        print("Kanal ekleniyor")
        time.sleep(1)
        self.kanallarListesi.append(kanalİsmi)
        print("Kanal Eklendi.")
    
    def rastgeleKanal(self):

        channel = random.randint(0,len(self.kanallarListesi)-1)

        self.kanal = self.kanallarListesi[channel]

        print(f"Suan {self.kanal} kanalını izliyorsunuz.")
    
    def _len_(self):
        return len(self.kanallarListesi)
    
    def __str__(self):
        return (f"Tv Durum :{self.tvDurum} \nTv Ses: {self.ses}\n İzlediğiniz Kanal: {self.kanal}\n Kanal Listesi:{self.kanallarListesi}\n ")

kumanda = commander()


print("""
Telezivyon kumanda uygulaması;
1.Tv aç
2.Tv Kapat
3.Ses Ayarı
4.Yeni Kanal Ekle
5. Kanal sayısını göster
6.Random Kanal Seç
7.Ürün Bilgileri
""")


while True:
    print("Programı sonlandırmak için q ya basınız")

    islem = input("İşlem seciniz:")

    if islem == "q":
        break
    elif islem == "1":
        kumanda.tvAcmak
    elif islem == "2":
        kumanda.tvKapat
    elif islem == "3":
        kumanda.sesAyar
    elif islem == "4":
        kanalİsimleri = input("Eklediginiz kanal ismini giriniz")
        kanalListesi = kanalİsimleri.split(",")
        for a in kanalListesi:
            kumanda.ekle(a)
    elif islem == "6":
        kumanda.rastgeleKanal()
    elif islem == "7":
        print(kumanda)
    elif(islem == "5"):
        print("Kanal sayısı", len(kumanda)
        
    else:
        print("GECERSİZ İSLEM")



```
