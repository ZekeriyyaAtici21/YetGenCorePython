# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python

import time 

def akihesap(kg,boy):
    aki = kg/(boy**2)

    return round(aki,1)

def aralik(aki):
    if aki < 18.5:
        return("Zayıf")
    elif aki <= 24.9:
        return("Normal")
    elif aki <= 29.9:
        return("Fazla Kilolu")
    elif aki <=39.9:
        return("Obez")
    elif aki == 40:
        return(" İleri Obez")

def cikislar(isim, yas, kilo, boy, aki, deger):
    kayitTarihi = time.ctime(time.time())
    t1 = f"Kullanıcı adı  {isim}|n"
    t2 = f"Yas, Kilo, Boy : {yas} {kilo} {boy}\n "
    t3 = f"Vücut kitle endeks : {aki}"
    t4 = f"deger sonucları : {deger}"
    t5 = f"Kayit tarihi {kayitTarihi}"
    return (t1+t2+t3+t4+t5)

def main():
    girisNumber = int(input(" Kullanıcı sayısını girin:  "))

    for x in range(girisNumber):
        print(f"Kullanıcı kaydı ", x+1)

        isim = input("Kullanıcı adı ve soyadı")
        yas = int(input("Kullanıcı yaşı"))
        kilo = float(input("Kullanıcı kilosı"))
        boy = float(input("Kullanıcı boyu"))

        aki = akihesap(kilo, boy)

        deger = aralik(aki)

        cikis = cikislar(isim,yas,kilo,boy,aki,deger)

        with open("saglik.txt", "a") as f :
            f.write(cikis)
            f.write(" "*15)
            f.write("\n")
        
    print("Veriler dosyaya kaydedildi.")

main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time 

menu = {"elma": 8.50,
        "armut": 10.00,
        "mandalin": 6.50,
        "portakal" : 8.00,
        "çilek" : 11.00,
        "uzum" : 6.00,
        "Karpuz": 9.00,
        "Kavun": 9.00
        }

def menuDisplay():
    print("-- YETBAR MENU --")

    for item in list(menu.keys()):

        print(f"{item.upper():<10} fiyatı : {menu[item]} TL")

def tekUrunHesaplama(urun, adet):

    assert adet >= 0 , " Ürün adedi 0 ve 0 dan küçük olamaz"

    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet
    

def satinAl():
    menuDisplay()

    print(" Satın almak istediğiniz ürünü giriniz. Çıkıs yapmak için '0' a basınız")

    alinanUrunList = {}
    alinanUrunAdeti = {}
    son_fiyat = 0

    while 1:
        eklenecek_urun = input("İstediginiz urunun adını giriniz")

        if eklenecek_urun == "0":
            print("Fiş kaydedildi")
            break

        ekleme_adedi = int(input("Kaç adet almak istediğinizi giriniz"))

        fiyat = tekUrunHesaplama(eklenecek_urun,ekleme_adedi)

        if fiyat == "None":
            print("Ürün eklenemedi. Lütfen başka bir ürün seçiniz.")
        else:
            print(f"Toplam Borcunuz: {fiyat}\n Satın almayı onaylamak için 'y' iptal etmek için 'n' basıınız. ")

            check = input("Satın almayı onaylıyor musunuz?  y or n")

            if check == "y":
                print("Satın alma onaylandı")
                print(" "*15)

                son_fiyat += fiyat

                alinanUrunList.append(eklenecek_urun)
                alinanUrunAdeti.append(ekleme_adedi)
                print(f"Ara toplam = {son_fiyat}")

            elif check == "n":
                print("Satın alma işlemi iptal edildi.")
                print(" "*20)
    return son_fiyat, alinanUrunList, alinanUrunAdeti


def main():
    print("YetBar a Hoigeldiniz")
    program = None

    while 1:
        program = int(input("Menüyü görmek için 1'e, smoothie yapmak için 2'ye, çıkmak içinde 0'a basiniz"))

        if program < 0 or program > 2:
            print("Yanlıs bir tercih yaptınız lütfen yeniden deneyiniz.")
        elif program == 0:
            print("Çikis yapilyor... \n")
            break
        elif program == 1:
            menuDisplay()
        elif program == 2:
            satin_alma_ucreti, alinanUrunList, alinanUrunAdeti = satinAl()
            with open("kasa.txt", "a") as f:
                
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(",".join(map(str,alinanUrunList)))
                f.write("\n")
                f.write(",".join(map(str,alinanUrunAdeti)))
                f.write("\n")
                f.write(",".join(map(str,satin_alma_ucreti)))
                f.write("\n")
                f.write(" "*20)

                cikis_kontrol = input("Başka bir işlem yapmak istiyor musunuz? , yes/no")

                if cikis_kontrol == "yes":
                    pass
                else:
                    print("YetBar tercih ettiğiniz için teşekkür ederiz")
                    break

main()
```
